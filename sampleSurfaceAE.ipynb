```python
import utils
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
decay_steps = [25,20,15,10]
decay_rate = [0.997,0.995,0.993,0.99]
current_iter = tf.Variable(0)     #for weight decay
for i in decay_rate:
    for j in decay_steps:
        X, optimizer, cost = utils.getNet(i,j,current_iter)
        total_cost = utils.trainAE("surface_train.tfrecords", X, optimizer, cost,current_iter)
        plt.plot(total_cost)
        plt.title('AutoEncoder Cost Change');plt.xlabel('Epochs');plt.ylabel("Cost")
        plt.savefig('decay_rate_%.5f_decay_steps_%d.png'%(i,j))
        plt.show()
```
